```python
#Import all the needed Packages
import numpy as np
import pandas as pd
import json
import seaborn as sns
import matplotlib.pyplot as plt
import sqlite3
from pandas import Series, DataFrame
```

```python
#Read csv Dataset
dframe1=pd.read_csv('C:/Users/subra/Desktop/Temp/ASS 5/HateCrimes_csv.csv')
dframe1.head()
```

```python
#Read json Dataset
dframe2 = pd.read_json('C:/Users/subra/Desktop/Temp/ASS 5/HateCrimes_json.json')
dframe2.head()
```

```python
#Read sql Dataset
con = sqlite3.connect('C:/Users/subra/Desktop/Temp/ASS 5/hatecrimes.db')
cur = con.cursor()
```

```python
df = pd.read_sql_query("SELECT * FROM crimes", con)
df
```

```python
#Modify column names to Merge datasets
dframe3=df.rename(index=str,columns={'state':'State','vRace':'Race','vRel':'Rel','vSexOr':'Sex','agName':'Agency_name','agType':'Agency_type'})
dframe3.head()
```

```python
#Merge first two Dataframes
m1=pd.merge(dframe1,dframe2,on=['Agency_name','State','Agency_type'],how='outer')
m1
```

```python
#Merge third Dataframe, create final Dataset and clean the dataset
merged=pd.merge(m1,dframe3,on=['Agency_name','State','Agency_type'],how='outer')
merged=merged.replace(np.NaN,0)
merged
```

```python
#Create a csv file
merged.to_csv("Final_Data.csv",index=False)
```

```python
#Describe the Dataset
merged.describe()
```

```python
#Create a new column which consists total no.of.crimes
merged['No.of.Crimes']=merged['quarter_1']+merged['quarter_2']+merged['quarter_3']+merged['quarter_4']
merged
```

```python
#Total No.of.Crimes
y=merged['No.of.Crimes'].sum()
y
```

```python
##Count of Annual crime for each Quarter
Crime_Sum = merged.groupby(['Agency_type']).agg({'quarter_1': 'sum','quarter_2': 'sum','quarter_3': 'sum','quarter_4': 'sum'})
Crime_Sum
```

```python
#Percentage of Annual crime for each Quarter
Crime_Percent = Crime_Sum.groupby(level=['Agency_type']).apply(lambda x:
                               100 * x / y)
Crime_Percent = round(Crime_Percent, 2)
Crime_Percent
```

```python
#Maximum number of crimes were reported by the agencies during which quarter of 2013
Q1 = merged['quarter_1'].sum()
Q2 = merged['quarter_2'].sum()
Q3 = merged['quarter_3'].sum()
Q4 = merged['quarter_4'].sum()
print ("Quarter1",Q1)
print ("Quarter2",Q2)
print ("Quarter3",Q3)
print ("Quarter4",Q4)
data={'No.of.Crime':[Q1,Q2,Q3,Q4]}
Max_No_of_Crime = pd.DataFrame(data, index = ['Quarter 1', 'Quarter 2', 'Quarter 3', 'Quarter 4']).sort_values(by='No.of.Crime',ascending=False)
Max_No_of_Crime
```

```python
#Agency type reported maximum number of crimes in 2013
Max_Crimes_Agency=merged['No.of.Crimes'].groupby(merged['Agency_type'])
Max_Crimes_Agency.sum()
```

```python
#States in which crimes have been committed against most ethnic groups?
Ethnicity=merged['Ethnicity'].groupby(merged['State'])
Ethnicity
```

```python
Ethnic_Groups=pd.DataFrame(Ethnicity.sum())
Ethnic_Groups=Ethnic_Groups.sort_values(by='Ethnicity',ascending=False)
Ethnic_Groups
```

```python
#Finding Outliers in Ethnicity
merged['Disability'].value_counts()
```

```python
#Removing Outliers
Disability_set=merged[merged["Disability"] < 2.0]
Disability_set['Disability'].value_counts()
```

```python
#Total No.of.Crimes
z=Disability_set['No.of.Crimes'].sum()
z
```

```python
#The count of crimes committed against disabled and non-disabled groups in each of the four quarters.
Disable_set = Disability_set.groupby(['Disability']).agg({'quarter_1': 'sum','quarter_2': 'sum','quarter_3': 'sum','quarter_4': 'sum'})
Disable_set
```

```python
#The percentage of crimes committed against disabled and non-disabled groups in each of the four quarters.
Disable_Percent = Disable_set.groupby(level=['Disability']).apply(lambda x:
                               100 * x / z)
Disable_Percent=round(Disable_Percent,2)
Disable_Percent
```
